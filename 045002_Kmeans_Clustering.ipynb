```python
# Define a function to identify outliers using Z-score
def detect_outliers_zscore(data, threshold=3):
    z_scores = (data - data.mean()) / data.std()
    outliers = (z_scores > threshold) | (z_scores < -threshold)
    return outliers

# Number of outliers in each non-categorical variable using Z-score method
outliers_pizza_id = detect_outliers_zscore(df['pizza_id']).sum()
outliers_order_id = detect_outliers_zscore(df['order_id']).sum()
outliers_quantity = detect_outliers_zscore(df['quantity']).sum()
outliers_unit_price = detect_outliers_zscore(df['unit_price']).sum()
outliers_total_price = detect_outliers_zscore(df['total_price']).sum()

# Print the number of outliers
print("Number of outliers in pizza_id:", outliers_pizza_id)
print("Number of outliers in order_id:", outliers_order_id)
print("Number of outliers in quantity:", outliers_quantity)
print("Number of outliers in unit_price:", outliers_unit_price)
print("Number of outliers in total_price:", outliers_total_price)
```

```python
from sklearn.cluster import KMeans
import time
import psutil  # For memory usage tracking
import os
import numpy as np  # Import NumPy for generating example data

# Function to measure memory usage
def memory_usage():
    process = psutil.Process(os.getpid())
    mem = process.memory_info().rss / 1024 / 1024  # Memory usage in MB
    return mem

# Function to perform K-Means clustering and measure time
def kmeans_time_statistics(data, n_clusters):
    start_time = time.time()

    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(data)

    end_time = time.time()
    execution_time = end_time - start_time

    memory_used = memory_usage()

    return execution_time, memory_used

# Example usage
data = np.random.rand(1000, 10)  # Example data, replace with your actual data
n_clusters = 6  # Number of clusters

execution_time, memory_used = kmeans_time_statistics(data, n_clusters)
print("Execution Time:", execution_time, "seconds")
print("Memory Used:", memory_used, "MB")
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from scipy.stats import chi2_contingency, f_oneway


# Define the categorical and non-categorical variables
categorical_features = ['pizza_name_id', 'pizza_size', 'pizza_category']
non_categorical_features = ['unit_price', 'total_price']

# Cluster Analysis with DBSCAN for Categorical Variables
def cluster_analysis_categorical(df, categorical_features):
    for feature in categorical_features:
        # Perform Chi-Square Test of Independence
        contingency_table = pd.crosstab(df['cluster_labels'], df[feature])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        print(f"Chi-Square Test for {feature}:")
        print(f"Chi-Square: {chi2}")
        print(f"P-value: {p}\n")

# Cluster Analysis with DBSCAN for Non-Categorical Variables
def cluster_analysis_non_categorical(df, non_categorical_features):
    for feature in non_categorical_features:
        # Perform ANOVA
        groups = [df[df['cluster_labels'] == label][feature] for label in df['cluster_labels'].unique()]
        f_statistic, p_value = f_oneway(*groups)
        print(f"ANOVA for {feature}:")
        print(f"F-statistic: {f_statistic}")
        print(f"P-value: {p_value}\n")

# Perform cluster analysis with DBSCAN and evaluate performance
def cluster_analysis_dbscan(df, categorical_features, non_categorical_features):
    # Fit DBSCAN model
    dbscan_model = DBSCAN(eps=0.5, min_samples=5)
    df['cluster_labels'] = dbscan_model.fit_predict(df[non_categorical_features])

    # Cluster analysis for categorical features
    cluster_analysis_categorical(df, categorical_features)

    # Cluster analysis for non-categorical features
    cluster_analysis_non_categorical(df, non_categorical_features)

# Perform cluster analysis with DBSCAN
cluster_analysis_dbscan(df, categorical_features, non_categorical_features)
```
